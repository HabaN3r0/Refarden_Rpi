```python
import cv2
import os
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
foldername = "C:/Users/user/Pictures/"
filename = "images.jpg"

img = cv2.imread(os.path.join(foldername, filename))

res = cv2.resize(img, dsize = (250, 250))
sat = cv2.cvtColor(res, cv2.COLOR_BGR2HSV)
```

```python
fig = plt.figure()

i1 = fig.add_subplot(2,2,1)
i1.imshow(sat)
i2 = fig.add_subplot(2,2,2)
i2.imshow(res)
```
